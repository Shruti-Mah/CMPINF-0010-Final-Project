# The Best Neighborhood

## Lowest Number of Felonies

```python
import pandas as pd
import matplotlib.pyplot as plt

#load file
chart = pd.read_csv("Police_Arrests_24_25.csv")

felonyCounts = chart[chart["ArrestCharge_Felony_Misdemeanor_Description"] == "FELONY"]["Neighborhood"].value_counts().reset_index()
felonyCounts.columns = ["Neighborhood", "FelonyCount"]
felonyCounts = felonyCounts.sort_values("FelonyCount", ascending = False)

topTen = felonyCounts.head(10)

plt.figure(figsize = (12,6))
plt.bar(topTen["Neighborhood"], topTen["FelonyCount"])
plt.xticks(rotation = 90)
plt.xlabel("Neighborhood")
plt.ylabel("Number of Felony Arrests")
plt.title("Top Ten Neighborhoods with Most Felony Arrests")
plt.tight_layout()
plt.show()

bottomTen = felonyCounts.tail(10).sort_values("FelonyCount")

least_neighborhood = bottomTen.iloc[0]["Neighborhood"]

print("\nBottom 10 Neighborhoods with Fewest Felonies: \n")
for i, row in bottomTen.iterrows():
    name = row["Neighborhood"]
    count = row["FelonyCount"]
    if name == least_neighborhood:
        print(f"*** {name}: {count} <-- Lowest***")
    else:
        print(f"{name}: {count}")

print("\nLowest felony count neighborhood: ", least_neighborhood)
```

## Lowest Number of Fire Incidents

```python
import pandas as pd
import matplotlib.pyplot as mp

fire_csv = "https://data.wprdc.org/datastore/dump/8d76ac6b-5ae8-4428-82a4-043130d17b02"
fireinc_data = pd.read_csv(fire_csv) #fire incidents

fire_df = fireinc_data[fireinc_data["type_description"] == "Building fire"]["neighborhood"].value_counts().reset_index()
fire_df.columns = ["Neighborhood", "IncidentCount"]
fire_df = fire_df.sort_values("IncidentCount", ascending = False)

leastIncidents = fire_df.iloc[-1]["Neighborhood"]

topTen = fire_df.head(10)

mp.figure(figsize = (12,6))
mp.bar(topTen["Neighborhood"], topTen["IncidentCount"], color = "teal")
mp.xticks(rotation = 90)
mp.xlabel("Neighborhood")
mp.ylabel("Number of Incidents")
mp.title("Top 10 Neighborhoods with Most Building Fires")
mp.tight_layout()
mp.show()

bottomTen = fire_df.tail(10).sort_values("IncidentCount")

print("\nBottom 10 Neighborhoods:\n")
for i, row in bottomTen.iterrows():
    n = row["Neighborhood"]
    c = row["IncidentCount"]
    if n == leastIncidents:
        print(f"*** {n}: {c} <-- Lowest ***")
    else: 
        print(f"{n}: {c}")

print("Neighborhood with least Building Fires: ", leastIncidents)
```

## Number of Disorderly Conducts

```python
import matplotlib.pyplot as mp
import pandas as pd

crimes_csv = "https://data.wprdc.org/datastore/dump/6b11e87d-1216-463d-bbd3-37460e539d86"
crime_data = pd.read_csv(crimes_csv)

crime_df = crime_data[["NEIGHBORHOOD", "OFFENSES"]]
query_mask = crime_df["OFFENSES"].astype(str).str.contains("5503", na = False)
crime_df = crime_df[query_mask]


crime_df = crime_df["NEIGHBORHOOD"].value_counts().reset_index()
crime_df.columns = ["Neighborhood","Disorderly Conducts"]
crime_df = crime_df.sort_values("Disorderly Conducts", ascending = False)

leastConducts = crime_df.iloc[-1]["Neighborhood"]

topTen = crime_df.head(10)
colors = ["red" if n == leastConducts else "black" for n in topTen["Neighborhood"]]

mp.figure(figsize = (12,6))
mp.bar(topTen["Neighborhood"], topTen["Disorderly Conducts"], color = colors)
mp.xticks(rotation = 90)
mp.xlabel("Neighborhood")
mp.ylabel("Number of Disorderly Conducts")
mp.title("Top 10 Neighborhoods with most Disorderly Conducts")
mp.tight_layout()
mp.show()

bottomTen = crime_df.tail(10).sort_values("Disorderly Conducts", ascending = True)

print("\nBottom 10 Neighborhoods with Least Disorderly Conducts: \n")
for i, row in bottomTen.iterrows():
    n = row["Neighborhood"]
    c = row["Disorderly Conducts"]
    if n == leastConducts:
        print(f"*** {n}: {c} <-- Lowest ***")
    else: 
        print(f"{n}: {c}")

print("Neighborhood with least Disorderly Conducts: ", leastConducts)
```

For the neighborhood with the lowest number of felonies it was the Strip District, for lowest fire incidents it was New Homestead, and for lowest number of disorderly conducts it was Ridgemont. However, none of these neighborhoods overlap or are the same. After comparing all of the charts and the neighborhoods in each we came to the conclusion that Arlington Heights is the safest neighborhood. It ranks either at 0 or close to 0 on every chart consistently. While other neighborhoods rank low in one category but high in another, Arlington Heights is low in every submetric that we have.
