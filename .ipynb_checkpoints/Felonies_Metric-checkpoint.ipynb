## Metric

Our overall metric is safety. There are a few features that come into play with this. One aspect that we chose to measure was the amount of felonies tied to a neighborhood. Using the dataset: [Pittsburgh Police Arrests](https://data.wprdc.org/dataset/pbp_arrest_data_2024_2025/resource/e419c20c-8df4-4729-830c-e49427a656e0). This dataset features all arrests from 2024 to 2025, and the categories we focused on: arrest type and neighborhood it took place. 

## The Best Neighborhood

```python
import pandas as pd
import matplotlib.pyplot as plt

#load file
chart = pd.read_csv("Police_Arrests_24_25.csv")

felonyCounts = chart[chart["ArrestCharge_Felony_Misdemeanor_Description"] == "FELONY"]["Neighborhood"].value_counts().reset_index()
felonyCounts.columns = ["Neighborhood", "FelonyCount"]
felonyCounts = felonyCounts.sort_values("FelonyCount", ascending = False)

topTen = felonyCounts.head(10)

plt.figure(figsize = (12,6))
plt.bar(topTen["Neighborhood"], topTen["FelonyCount"])
plt.xticks(rotation = 90)
plt.xlabel("Neighborhood")
plt.ylabel("Number of Felony Arrests")
plt.title("Top Ten Neighborhoods with Most Felony Arrests")
plt.tight_layout()
plt.show()

bottomTen = felonyCounts.tail(10).sort_values("FelonyCount")

least_neighborhood = bottomTen.iloc[0]["Neighborhood"]

print("\nBottom 10 Neighborhoods with Fewest Felonies: \n")
for i, row in bottomTen.iterrows():
    name = row["Neighborhood"]
    count = row["FelonyCount"]
    if name == least_neighborhood:
        print(f"*** {name}: {count} <-- Lowest***")
    else:
        print(f"{name}: {count}")

print("\nLowest felony count neighborhood: ", least_neighborhood)
```

Using the bar chart above we can see that the Strip District has the fewest number of felonies. This count is from 2024 to last month of 2025, which is a substantial period of time. Seeing as this neighborhood has gone the longest without a felony, it can be labelled as "The Best Neighborhood" using this metric! 

## Conclusion

There are several neighborhoods that might equal the Strip District in number of felonies, the metric simply went off how they were listed in the csv file. But seeing as the neighborhood does have only 1 felony, we can conclude that this neighborhood is the best one to live in based off of that number.
