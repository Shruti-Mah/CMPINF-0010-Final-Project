## **Introduction**

We are attempting to find the best neighborhood in Pittsburgh. Our overall metric is safety, because safety is very important in day-to-day-life.

#### **Metric**

Our overall metric is safety. The specific submetric for this dataset is the number of disorderly conducts in each neighborhood in Pittsburgh. We are looking for the neighborhood with the least amount of disorderly conduct citations, because that would mean it is safer to be in.

## **The Best Neighborhood**

```python
import pandas as pd

crimes_csv = "https://data.wprdc.org/datastore/dump/6b11e87d-1216-463d-bbd3-37460e539d86"
crime_data = pd.read_csv(crimes_csv)

crime_data.head(10)
```

```python
crime_df = crime_data[["NEIGHBORHOOD", "OFFENSES"]]
query_mask = crime_df["OFFENSES"].str.contains("5503") # disorderly conduct code
crime_df = crime_df[query_mask]
crime_df = crime_df.sort_values("NEIGHBORHOOD", ascending=False)
crime_df.head(10)
```

```python
import matplotlib.pyplot as mp
crime_df = crime_df["NEIGHBORHOOD"].value_counts()
crime_df
```

```python
crime_df = crime_data[crime_data["OFFENSES"].str.contains("5503")]["NEIGHBORHOOD"].value_counts().reset_index()
crime_df.columns = ["Neighborhood","Disorderly Conducts"]
crime_df = crime_df.sort_values("Disorderly Conducts", ascending = False)

leastConducts = crime_df.iloc[-1]["Neighborhood"]

topTen = crime_df.head(10)
colors = ["red" if n == leastConducts else "black" for n in topTen["Neighborhood"]]

mp.figure(figsize = (12,6))
mp.bar(topTen["Neighborhood"], topTen["Disorderly Conducts"], color = colors)
mp.xticks(rotation = 90)
mp.xlabel("Neighborhood")
mp.ylabel("Number of Disorderly Conducts")
mp.title("Top 10 Neighborhoods with most Disorderly Conducts")
mp.tight_layout()
mp.show()

bottomTen = crime_df.tail(10).sort_values("Disorderly Conducts", ascending = True)

print("\nBottom 10 Neighborhoods with Least Disorderly Conducts: \n")
for i, row in bottomTen.iterrows():
    n = row["Neighborhood"]
    c = row["Disorderly Conducts"]
    if n == leastConducts:
        print(f"*** {n}: {c} <-- Lowest ***")
    else: 
        print(f"{n}: {c}")

print("Neighborhood with least Disorderly Conducts: ", leastConducts)
```

## **Conclusion**
The neighborhood with the least amount of disorderly conducts was Ridgemont.
